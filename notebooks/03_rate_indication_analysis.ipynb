```python
import pandas as pd

df = pd.read_csv("../data/insurance_pricing_dataset.csv")

target_loss_ratio = 0.65

segment = df.groupby("vehicle_type").agg(
    policies=("claim_flag", "count"),
    total_premium=("current_premium", "sum"),
    total_claims=("claim_amount", "sum"),
    claim_count=("claim_flag", "sum")
).reset_index()

segment["loss_ratio"] = segment["total_claims"] / segment["total_premium"]
segment["claim_frequency"] = segment["claim_count"] / segment["policies"]
segment["indicated_rate_change"] = segment["loss_ratio"] / target_loss_ratio - 1

print(segment.sort_values("indicated_rate_change", ascending=False))
```
